```python
# here we see there are too many outliers so we can use 1.5IQR method to remove outliers
for col in cols:
    q1=new_df[col].quantile(0.25)
    q3=new_df[col].quantile(0.75)
    iqr = (q3-q1)
    upper_range = q3+(1.5*iqr)
    lower_range = q1-(1.5*iqr)
    new_df=new_df.loc[(new_df[col] > lower_range) & (new_df[col] < upper_range)]
```
